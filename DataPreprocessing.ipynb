```python
import pandas as pd
import numpy as np

# Sample dataset with missing values
data = {
    "Name": ["Ali", "Sara", "Hassan", "Ayesha", np.nan, "Zain", "Fatima", np.nan, "Bilal", "Maryam"],
    "Age": [25, 30, np.nan, 22, 28, np.nan, 26, 29, 31, np.nan],
    "City": ["Lahore", np.nan, "Karachi", "Islamabad", "Lahore", "Karachi", np.nan, "Islamabad", "Lahore", "Karachi"],
    "Salary": [50000, 60000, 55000, np.nan, 58000, 62000, 57000, np.nan, 61000, 59000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.isnull().sum()
# df.dropna()
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
print(df)
df.dropna()
df.dtypes
```
